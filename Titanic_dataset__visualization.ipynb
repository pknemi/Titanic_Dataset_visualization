```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Titanic-Dataset.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
data["Age"].mean()
```

```python
data["Age"].isnull().sum()
```

```python
data["Cabin"].isnull().sum()
```

```python
data["Cabin"].mode()
```

```python
data["Cabin"].fillna(data["Cabin"].mode(),inplace=True)
```

```python
data["Embarked"].fillna(data["Embarked"].mode(),inplace=True)
```

```python
data.head(10)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
sns.boxplot(y=data["Age"])
```

```python
data["Age"].fillna(data["Age"].median(),inplace=True)
```

null values in "age" column is treated with median as it contains outlier.

```python
data["Cabin"].fillna(data["Cabin"].mode()[0],inplace=True)
```

null values in cabin column is replaced with mode


```python
data["Cabin"].fillna(method="ffill",inplace=True)
```

```python
data["Embarked"].fillna(data["Embarked"].mode(),inplace=True)
```

null values in embarked is also treated with mode.

```python
plt.style.available
```

## **Countplot for Survival Based on Gender**

```python
plt.style.use('dark_background')
```

```python
sns.countplot(x='Survived', hue='Sex', palette="pastel", data=data)
plt.title('Survival Count by Gender')
plt.show()
```

Insight:around 350 passengers have survived out of which more than 200 is female.

# **Histogram of Age Distribution**


```python
plt.style.use('dark_background')
plt.hist(data['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Passengers')
plt.show()
```

Insight:Most of the titanic passengers comes under the age group of 25-30

## **Boxplot of Fare by Passenger Class**

```python
sns.boxplot(x='Pclass', y='Fare', data=data)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

Insight:This show the distribution of fares paid by passengers in different classes.Each class contain outliers

```python
sns.countplot(x='Survived', hue='Pclass', palette="muted", data=data)
plt.title('Survival Count by Passenger Class')
plt.show()
```

Insight:Most of the passengers who are not survived comes under  third class.

# **Distribution of Passenger Classes**

```python
class_counts = data['Pclass'].value_counts()
plt.pie(class_counts, labels=['Class 1', 'Class 2', 'Class 3'], autopct='%1.1f%%')
plt.title('Distribution of Passenger Classes')
plt.show()
```

Insight:The pie chart shows that  55% of total passengers comes under first class.



# **Scatterplot of Age vs Fare, Colored by Survival**

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data)
plt.title('Age vs Fare, Colored by Survival')
plt.show()
```

Insight:most of the survived passengers comes under the age group 20-40.


## **Violinplot of Age by Survival and Gender**

```python
sns.violinplot(x='Survived', y='Age', hue='Sex', data=data, split=True)
plt.title('Age Distribution by Survival and Gender')
plt.show()
```

It is analysed that most number of male and female passengers survived are of the age group 20-40


# **Pairplot of Selected Features**


```python
sns.pairplot(data[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.title('Pairplot of Selected Features')
plt.show()
```

# **Passengers per Embarkation Port**

```python
sns.countplot(x='Embarked', data=data)
plt.title('Passengers per Embarkation Port')
plt.show()
```

Most number of passengers both survived and not survived are embarked from the port named "S".

## **Age Distribution Over Passenger Classes**

```python
age_by_class = data.groupby('Pclass')['Age'].mean()
plt.plot(age_by_class.index, age_by_class.values, marker='o')
plt.xlabel('Passenger Class')
plt.ylabel('Average Age')
plt.title('Average Age by Passenger Class')
plt.xticks([1, 2, 3])
plt.show()
```

Insight:The analysis reveals that the average age of passengers varies by class, with those in Class 1 having an average age between 38 and 40 years, while passengers in Class 2 average 30 years, and those in Class 3 are younger, with an average age of 28 years.

## **Proportion of Survivors and Non-Survivors:**




```python
survival_counts = data['Survived'].value_counts()
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%')
plt.title('Proportion of Survivors and Non-Survivors')
plt.show()
```

Insight:It can be seen that only 38.4% of the total passengers were survived.
